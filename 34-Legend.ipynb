```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days  = [1,2,3,4,5,6,7,]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.plot(days, max_t)
```

```python
# for plotting all in the same chart.
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days, max_t, label="max_temp")
plt.plot(days, min_t, label="min_temp")
plt.plot(days, avg_t, label="avg_temp")
#not printing the labels ?
```

```python
# for plotting all in the same chart.
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days, max_t, label="max_temp")
plt.plot(days, min_t, label="min_temp")
plt.plot(days, avg_t, label="avg_temp")

plt.legend()
```

```python
# say you want legend in upper right hand corner. 
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days, max_t, label="max_temp")
plt.plot(days, min_t, label="min_temp")
plt.plot(days, avg_t, label="avg_temp")

plt.legend(loc = "upper right")
```

```python
# you can also set it to default, it chooses best place. 
# say you want legend in upper right hand corner. 
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days, max_t, label="max_temp")
plt.plot(days, min_t, label="min_temp")
plt.plot(days, avg_t, label="avg_temp")

plt.legend(loc = "best")
```
