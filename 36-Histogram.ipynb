```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar)

# we notice that by default there are 10bin. 10 ranges. 
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, bins = 3, rwidth = .9)
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, bins = [80,100,125,150], rwidth = .9, color = 'g')
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, bins = [80,100,125,150], rwidth = .9, histtype = 'step')
```

```python
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.hist(blood_sugar_women, bins = [80, 100, 125, 150], rwidth = .99)
plt.hist(blood_sugar, bins = [80, 100, 125, 150], rwidth = .95)
```

```python
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.hist([blood_sugar_women, blood_sugar], 
         bins = [80, 100, 125, 150], 
         rwidth = .95, 
         label = ["women", "men"])
plt.legend()
plt.xlabel("sugar range")
plt.ylabel("Number of patients")
plt.title("Blood Sugar Analysis")
```

```python
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.hist([blood_sugar_women, blood_sugar], 
         bins = [80, 100, 125, 150], 
         rwidth = .95, 
         orientation = "horizontal",
         label = ["women", "men"])
plt.legend()
plt.xlabel("sugar range")
plt.ylabel("Number of patients")
plt.title("Blood Sugar Analysis")
```
