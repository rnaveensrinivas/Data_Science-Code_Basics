```python
df = pd.read_csv("weather_data2.csv", parse_dates = ['day'])
df
```

```python
type(df.day[0])
```

```python
# now we want day to be the index. 
df.set_index("day", inplace = True)
df
```

```python
new_df = df.fillna(0) # replacing all nan with 0.
```

```python
new_df
```

```python
# can we do specific for each column?
new_df = df.fillna({
    'temperature': 0, 
    'windspeed': 0,
    'event': 'no event'
})
new_df
```

```python
# can we do better? other than 0? It will affect mean calculation.
# carry forward?
 
new_df = df.fillna(method="ffill") # ffill means forward fill. 
new_df
```

```python
new_df = df.fillna(method="bfill") # bfill means backward fill. 
new_df
```

```python
# so far we were copying vertically. But can we do horizontally?
new_df = df.fillna(method="bfill", axis="columns")
new_df
```

```python
# putting a restrcition on how much we can carry forward!
new_df = df.fillna(method="ffill", limit=1 )
new_df
```

```python
# Again can we do better? like the numbers so unrealistic. 
# can we make better guess?
# we will use interpolate. 
# it will not work for textual data.

new_df = df.interpolate()
new_df
```

```python
# again let's look at day and compare temperature with it. 
# not so great. 
# so will make the interpolation depend on day. 

new_df = df.interpolate(method="time")
new_df
```

```python
new_df = df.interpolate(method="day")
new_df
```

```python
# say you don't want all this, you just want to drop all the values with nan.

new_df = df.dropna() # drops if the row has at least one nan.
new_df
```

```python
# say you want to drop the rows if they have all empty cell.
new_df = df.dropna(how="all")
new_df
```

```python
# let's say you want a certin number of valid cells in a row
# this can be achieved using thresh

new_df = df.dropna(thresh=2)
new_df
```

```python
new_df = df.dropna(thresh=5)
new_df
```

```python
new_df = df.dropna(thresh=3)
new_df
```
