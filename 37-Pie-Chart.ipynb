```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400, 600, 300, 410, 750]
exp_labels = ["Home rent", "food", "Phone", "Car", "others"]
```

```python
plt.pie(exp_vals, labels = exp_labels)
```

```python
# to remove all the unwanted text use
plt.pie(exp_vals, labels = exp_labels)
plt.show()
```

```python
# to remove all the unwanted text use
plt.pie(exp_vals, labels = exp_labels, radius = 2)
plt.show()
```

```python
plt.pie(exp_vals, labels = exp_labels, autopct="%0.2f%%")
plt.show()
```

```python
plt.pie?
```

```python
# now you want a particular pie out. 
plt.pie(exp_vals, labels = exp_labels, autopct="%0.2f%%", 
        explode = [0, 0.5, 0, .1, 0])
plt.show()
```

```python
# start angle gives at what angle the chart should start. 
```

```python
plt.pie(exp_vals, labels = exp_labels, autopct="%0.2f%%", startangle = 180)
plt.show()
```
