```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
company = ['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue = [90, 136, 89, 27]
```

```python
plt.xlabel('revenue')
plt.ylabel('company')
plt.bar(revenue, company)
```

```python
plt.xlabel('company')
plt.ylabel('revenue')
plt.bar(company, revenue)
plt.title("US Tech Stocks")
```

```python
plt.xlabel('company')
plt.ylabel('Billion dollars')
plt.bar(company, revenue, label = "revenue")
plt.title("US Tech Stocks")
plt.legend()
```

```python
profit = [40,2,34,12]
plt.xlabel('company')
plt.ylabel('Billion dollars')
plt.bar(company, revenue, label = "revenue")
plt.bar(company, profit, label ="profit")
plt.title("US Tech Stocks")
plt.legend()
```

```python
profit = [40,2,34,12]
plt.ylabel('company')
plt.xlabel('Billion dollars')
plt.barh(company, revenue, label = "revenue")
plt.barh(company, profit, label ="profit")
plt.title("US Tech Stocks")
plt.legend()
```
