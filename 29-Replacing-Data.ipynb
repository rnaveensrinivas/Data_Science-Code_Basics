```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("weather_data3.csv")
df
```

```python
new_df = df.replace([-99999, -88888], np.nan) 
# supplying a list values that is to be replaced.
new_df
```

```python
new_df = df.replace({
    'temperature': [-99999, -88888],
    'windspeed': [-99999, -88888],
    'event': '0' # since type of event is str.
}, {
    'temperature': [100000000, 200000000],
    'windspeed': [100000000, 200000000],
    'event': '0000' # since type of event is str.
})
new_df
```

```python
# replacing anything we anything we want, but across all cells. 
new_df = df.replace({
    -99999: np.nan,
    '0': 'Sunny'
})
new_df
```

```python
# say if temperature, windspeed had units. We don't want that. 
new_df = df.replace('[A-Za-z]', '', regex=True)
new_df

# applied to all cells, but we want only for few columns. 
```

```python
new_df = df.replace({
    'temperature': '[A-Za-z]',
    'windspeed': '[A-Za-z]',
}, '', regex = True)
new_df
```

```python
# now, say you want to replace a list of values with another list of values.
df = pd.DataFrame({
    'score': ['exceptional','average', 'good', 'poor', 'average', 'exceptional'],
    'student': ['rob', 'maya', 'parthiv', 'tom', 'julian', 'erica']
})
df
```

```python
new_df = df.replace(["exceptional", "average", "good", "poor"], ["O", "A", "B", "C"])
new_df
```
