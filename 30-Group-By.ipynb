```python
import pandas as pd
```

```python
df = pd.read_csv("weather_data_cities.csv")
df
```

```python
# now we want answers to following questions.
# 1. Find max temp for each city. 
# 2. Find avg wind speed per city.
```

```python
g = df.groupby('city')
g # g is a datafram group by object. 
# same as SQL group by.
# SELECT * FROM df GROUP BY city;
```

```python
for city, city_df in g:
    print(city)
    print(city_df)
```

```python
g.get_group('mumbai')
```

```python
g.max()
# first we are breaking the data set in to groups. 
# then we are performing operations on these groups. 

# split - apply - combine.
```

```python
g.mean()
```
