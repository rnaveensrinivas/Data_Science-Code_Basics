```python
# JSON = Java Script Object Notation
# JSON is a data exchange format similar to XML
```

```python
# JSON is a lightweight format compared to XML. 
# XML uses tags which is heavy weight. 
```

```python
# Two programs: 
# 1. Create an address book and write some records to it. 
# 2. Read this address book. 
```

```python
book = {} 

book['tom'] = {
    'name': 'tom', 
    'address': '1 red street, NY', 
    'phone': 1234567890,
}

book['bob'] = {
    'name': 'bob', 
    'address': '1 green street, NY', 
    'phone': 1234567891,
}
```

```python
import json
s = json.dumps(book) # converts the dictionary object into a json string format. 
print(s)
```

```python
with open("C://Users//sri//Desktop//DS-Codebasics//book.txt", "w") as f:
    f.write(s)
```

```python
# Now you can read this JSON data using any language that supports it. 
# Hence it is called data exchange format. 
```

```python
f = open("C://Users//sri//Desktop//DS-Codebasics//book.txt", "r")
```

```python
s = f.read()
type(s)
```
