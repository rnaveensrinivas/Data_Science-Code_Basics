```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50, 51, 52, 48, 47, 49, 46]
```

```python
plt.plot(x, y)
```

```python
plt.plot(x,y,'gx') # 'gx' = color marker
```

```python
plt.plot(x,y,'go--') # 'g(o/x/+)--' = color marker and order doesn't matter.
```

```python
plt.plot(x,y, color = "red", marker = "D", linestyle = "", markersize = 10)
# colour can be in hexadecimal as well. 
```

```python
# another propery is alpha = for transparency of line. 
plt.plot(x, y, 'g+', alpha = 0.3)
```
