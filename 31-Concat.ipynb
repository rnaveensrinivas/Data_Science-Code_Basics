```python
import pandas as pd

india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
df = pd.concat([india_weather, us_weather]) 
df

# here it is using the index from original dataframe. 
```

```python
df.reset_index()
```

```python
df.drop(["index"], axis = 0)
df
```

```python
df = pd.concat([india_weather, us_weather], ignore_index = True) 
df
```

```python
df = pd.concat([india_weather, us_weather], keys = ["india", "us"]) 
df
```

```python
df.loc["india"]
# it is very useful when we have merged many dataframes. 
```

```python
temperature_df = pd.DataFrame({
    'city': ['mumbai', 'delhi', 'banglore'],
    'temperature': [32,45,30], 
})
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    'city': ['delhi', 'banglore'],
    'windspeed': [7,12,], 
})
windspeed_df
```

```python
# now we want to concat them. 

df = pd.concat([temperature_df, windspeed_df])
df

# not what we want. 
```

```python
# axis = 1

df = pd.concat([temperature_df, windspeed_df], axis = 1)
df

```

```python
s = pd.Series(["Rain"], name = "event")
s
```

```python
df = pd.concat([temperature_df, s], axis = 1)
df
```
